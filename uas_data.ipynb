```python
import pandas as pd

# Baca data utama
df = pd.read_csv('data_pwk.csv')

# Gabungkan data luas wilayah
luas_list = []
for tahun in [2019, 2020, 2021, 2022]:
    luas_df = pd.read_csv(f'Luas_Purwakarta_{tahun}.csv')
    luas_df['Tahun'] = tahun
    luas_list.append(luas_df[['Kecamatan', 'Tahun', 'Luas Wilayah (Km2)']])

# Gabungkan semua
semua_luas = pd.concat(luas_list)

# Merge dengan data utama
df = pd.merge(df, semua_luas, on=['Kecamatan', 'Tahun'])

# Pindahkan kolom Luas sebelum Tahun
cols = df.columns.tolist()
luas = cols.pop(cols.index('Luas Wilayah (Km2)'))
tahun_idx = cols.index('Tahun')
cols.insert(tahun_idx, luas)
df = df[cols]

# Simpan
df.to_csv('data_final.csv', index=False)
```

```python
df_desa_2019 = pd.read_csv('desa_2019.csv')
df_desa_2020 = pd.read_csv('desa_2020.csv')
df_desa_2021 = pd.read_csv('desa_2021.csv')
df_desa_2022 = pd.read_csv('desa_2022.csv')
```

```python
df_desa_2019['Tahun'] = 2019
df_desa_2020['Tahun'] = 2020
df_desa_2021['Tahun'] = 2021
df_desa_2022['Tahun'] = 2022
```

```python
df_desa_all = pd.concat([df_desa_2019, df_desa_2020, df_desa_2021, df_desa_2022], 
                        ignore_index=True)
```

```python
df_desa_all = df_desa_all.dropna(subset=['Kecamatan', 'Jumlah Desa'])
df_desa_all['Kecamatan'] = df_desa_all['Kecamatan'].astype(str).str.strip()
df_final['Kecamatan'] = df_final['Kecamatan'].astype(str).str.strip()
```

```python
df_desa_all = df_desa_all[~df_desa_all['Kecamatan'].str.contains('Catatan|^\s*$', na=False)]
```

```python
df_final = df_final.merge(
    df_desa_all[['Kecamatan', 'Jumlah Desa', 'Tahun']],
    on=['Kecamatan', 'Tahun'],
    how='left'
)
```

```python
cols = df_final.columns.tolist()
cols.insert(2, cols.pop(cols.index('Jumlah Desa')))  # Pindah ke index 2
df_final = df_final[cols]
```

```python
df_final.to_csv('data_final1.csv', index=False)
```
