```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('data_final1.csv')
```

```python
def kategori_kepadatan(nilai):
    if nilai < 500:
        return "Rendah"
    elif 500 <= nilai <= 1500:
        return "Sedang"
    else:
        return "Tinggi"
```

```python
df['Kategori_Kepadatan'] = df['Kepadatan Penduduk per km persegi Km2'].apply(kategori_kepadatan)
```

```python
encoding_map = {'Rendah': 0, 'Sedang': 1, 'Tinggi': 2}
df['Kategori_Kepadatan_Encoded'] = df['Kategori_Kepadatan'].map(encoding_map)
```

```python
features = ['Jumlah Penduduk (Ribu)','Laju Pertumbuhan Penduduk per Tahun', 'Jumlah Desa', 'Rasio Jenis Kelamin Penduduk']

X = df[features]
y = df['Kepadatan Penduduk per km persegi Km2'] 
```

```python
df
```

```python
# StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Menampilkan intercept regresi dari model
intercept = model.intercept_
intercept
```

```python
# Menampilkan koefisien regresi dari model
coefficients = model.coef_
coefficients
```

```python
# Menghitung MSE dan R2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```
